```python
#importing all dependancies
%matplotlib inline 
#the above command displays the plot within the application without explicitly calling the show() function
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#line plots

x=np.linspace(0,20,100)  #returns 100 evenly spaced numbers over the specified interval (0,20)
y=np.sin(x)

fig,ax=plt.subplots()
ax.plot(x,y)
ax.set(title="Simple Line Plot",
       xlabel="x-axis",
       ylabel="y-axis");
```

```python
#scatter plots

np.random.seed(3)
x=np.random.random(10) #returns 10 floats in the range (0.0,1.0)
y=np.random.random(10)
fig,ax=plt.subplots()
ax.scatter(x,y)
ax.set(title="Scatter Plot",
      xlabel="x-axis",
      ylabel="y-axis");
```

```python
#bar graphs 

data={"A":10,
     "B":4,
     "c":7,
     "D":3,
     "E":1,
     "F":5}
fig,ax=plt.subplots()
ax.bar(data.keys(),data.values())
ax.set(title="Bar Graph",
      xlabel="x-axis",
      ylabel="y-axis");
```

```python
#horizontal bar graphs 

data={"A":10,
     "B":4,
     "c":7,
     "D":3,
     "E":1,
     "F":5}
fig,ax=plt.subplots()
ax.barh(list(data.keys()),data.values()) #the y parameter must be in list format so we convert it
ax.set(title="Bar Graph",
      xlabel="x-axis",
      ylabel="y-axis");
```

```python
#histogram

np.random.seed(1)
x=np.random.randn(1000) #returns 1000 samples from the standard normal distribution
fig,ax=plt.subplots()
ax.hist(x)
ax.set(title="Histogram",
      xlabel="x-axis",
      ylabel="y-axis");
```

```python
#time series

import datetime, random
random.seed(1)
data=pd.DataFrame({'date': np.array([datetime.datetime(2022,11,i+1) for i in range(8)]),
                  'number': random.sample(range(1,20),8)})
fig,ax=plt.subplots()
ax.plot(data.date, data.number)
ax.set(title="Time Series",
      xlabel="Dates",
      ylabel="Number")
plt.xticks(rotation=30, ha='right');
```

```python
#pie chart

random.seed(3)
x=np.array(random.sample(range(1,100),7))
plt.pie(x);
```
